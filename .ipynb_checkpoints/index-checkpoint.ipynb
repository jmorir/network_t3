## Preparación de la data y los grafos

## Distancia

Se observa la distancia de los grafos entre sí. Como todos los nodos están conectados entre sí, la distancia mayor entre dos nodos y la distancia total son iguales.

```python
 nx.diameter(Alcaldes)
```

```python
largest_cc_nodes = max(nx.connected_components(Alcaldes), key=len)
Alcaldes_largest=Alcaldes.subgraph(largest_cc_nodes)

nx.diameter(Alcaldes_largest)
```

```python
# geodesics
geodesics = dict(nx.all_pairs_shortest_path_length(Alcaldes))

# as matrix
MatrixGeo = [[geodesics.get(n,{}).get(m)  for m in Alcaldes.nodes] for n in Alcaldes.nodes]

# as DF (pandas)
MatrixGeo_DF=pd.DataFrame(MatrixGeo,index=Alcaldes.nodes, columns=Alcaldes.nodes)

# we get
MatrixGeo_DF
```

```python
import numpy as np

pd.DataFrame(np.unique(MatrixGeo_DF.values, return_counts=True),index=['valor','freq']).T.set_index('valor')
```

```python
MatrixReach_DF=(MatrixGeo_DF>0).replace({True:1,False:0})
MatrixReach_DF
```

El índice de conectividad termina siendo 1 ya que todos están conectados entre sí.

```python
nNodes=len(Alcaldes.nodes)
MatrixReach_DF.sum().sum()/(nNodes*(nNodes-1))
```

```python
# geodesics
geodesics_Lcc = dict(nx.all_pairs_shortest_path_length(Alcaldes_largest))

# as matrix
MatrixGeo_Lcc = [[geodesics_Lcc.get(n,{}).get(m)  for m in Alcaldes_largest.nodes] for n in Alcaldes_largest.nodes]

# as DF (pandas)
MatrixGeo_DF_Lcc=pd.DataFrame(MatrixGeo_Lcc,index=Alcaldes_largest.nodes, columns=Alcaldes_largest.nodes)

# we get
MatrixGeo_DF_Lcc
```

```python
# reachability
MatrixReach_DF_Lcc=(MatrixGeo_DF_Lcc>0).replace({True:1,False:0})
MatrixReach_DF_Lcc
```

```python
nx.density(Alcaldes_largest)
```

```python
AlcaldeMeasures=pd.concat([pd.DataFrame.from_dict(nx.degree_centrality(Alcaldes_largest),
                                  orient='index',
                                  columns=['Degree']),
           pd.DataFrame.from_dict(nx.closeness_centrality(Alcaldes_largest),
                                  orient='index',
                                  columns=['Closeness']),
           pd.DataFrame.from_dict(nx.betweenness_centrality(Alcaldes_largest),
                                  orient='index',
                                  columns=['Betweenness'])],axis=1)
## data
AlcaldeMeasures
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10,10))

ax.scatter(AlcaldeMeasures.Closeness, AlcaldeMeasures.Degree,s=1000*AlcaldeMeasures.Betweenness,
           c=100*AlcaldeMeasures.Betweenness,
           cmap=plt.get_cmap('YlOrRd'), alpha=0.4)

valsForAnnotate=zip(AlcaldeMeasures.index,AlcaldeMeasures.Closeness,AlcaldeMeasures.Degree)
for name,coordX,coordY in valsForAnnotate:
    ax.annotate(name, (coordX,coordY),alpha=0.5)

plt.title("Centrality  (size and  color for Betweenness)", y=-0.1)
plt.xlabel("Closeness")
plt.ylabel("Degree")
plt.show()
```

## Graficación final

El siguiente gráfico fue realizado con el apoyo de lso recursos de ChatGPT. Lo realicé para poder presentar la información con mucho más cuidado y claridad. 

```python
#pip install dash plotly
```

```python
import networkx as nx
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.graph_objects as go

# Crear el grafo (suponiendo que tu grafo se llama 'Alcaldes')
G = Alcaldes

# Usar el layout spring para posicionar los nodos
pos = nx.spring_layout(G)

# Función para crear el gráfico
def create_graph(show_labels):
    # Obtener las posiciones de las aristas
    edge_x = []
    edge_y = []
    for edge in G.edges():
        x0, y0 = pos[edge[0]]
        x1, y1 = pos[edge[1]]
        edge_x.extend([x0, x1, None])
        edge_y.extend([y0, y1, None])

    # Crear el trazo para las aristas (líneas que conectan nodos)
    edge_trace = go.Scatter(
        x=edge_x, y=edge_y,
        line=dict(width=0.5, color='red', dash='dash'),  # Color y estilo de línea
        hoverinfo='none',
        mode='lines'
    )

    # Obtener las posiciones de los nodos
    node_x = []
    node_y = []
    for node in G.nodes():
        x, y = pos[node]
        node_x.append(x)
        node_y.append(y)

    # Crear el trazo para los nodos
    node_trace = go.Scatter(
        x=node_x, y=node_y,
        mode='markers',
        hoverinfo='none',  # No mostrar nombres al pasar el cursor
        marker=dict(
            color='yellow',  # Color de los nodos
            size=10,
            line_width=2),
    )

    # Crear el trazo para las etiquetas de los nodos (si se muestran)
    label_trace = go.Scatter(
        x=node_x, y=node_y,
        mode='text',
        text=list(G.nodes()),  # Etiquetas de los nodos
        textposition="bottom center",  # Posición de las etiquetas
        showlegend=False,
        visible='legendonly' if not show_labels else True  # Ocultar etiquetas si show_labels es False
    )

    # Crear la figura y configurarla como interactiva
    fig = go.Figure(data=[edge_trace, node_trace, label_trace],
                    layout=go.Layout(
                        title='<b>Red de Alcaldes distritales de Lima 2019-2022 y 2023-2026</b>',
                        titlefont_size=16,
                        showlegend=False,
                        hovermode='closest',
                        margin=dict(b=0, l=0, r=0, t=40),
                        xaxis=dict(showgrid=False, zeroline=False),
                        yaxis=dict(showgrid=False, zeroline=False))
                   )

    return fig

# Crear la aplicación Dash
app = dash.Dash(__name__)

# Layout de la aplicación
app.layout = html.Div([
    html.H1("Grafo Interactivo de Alcaldes"),
    dcc.Graph(id='graph', figure=create_graph(show_labels=True)),  # Por defecto, mostrar etiquetas
    html.Button("Mostrar/Ocultar", id='toggle-button', n_clicks=0)  # Botón para activar/desactivar nombres
])

# Callback para actualizar el gráfico
@app.callback(
    Output('graph', 'figure'),
    Input('toggle-button', 'n_clicks')
)
def update_graph(n_clicks):
    show_labels = n_clicks % 2 == 1  # Cambia la visibilidad de las etiquetas
    return create_graph(show_labels)

# Ejecutar la aplicación
if __name__ == '__main__':
    app.run_server(debug=False)
```

## Interpretación

Tras realizar la graficación, se puede observar que hay una relación muy cercana de los Alcaldes y se forman agrupaciones cercanas por partido político. En ese sentido, es posible observar 6 grandes clusters de nodos de los partidos con mayor representación en los gobiernos locales: Renovación Popular (antes Solidaridad Nacional), Alianza Para el Progreso, Acción Popular, Somos Perú y Podemos Perú. Debido a la gran cantidad de nodos y su densidad en estos subgrupos, podríamos decir que tienen gran presencia en Lima Metropolitana y sostenida durante las últimas dos elecciones. Por el contrario, hay nodos que se encuentran muy alejados y pertenecen a alcaldes de partidos que sólo ganaron una elección o que fueron creados para competir en el distrito (como San Miguel Me Gusta). En ese sentido, la mayor parte de los nodos son centrales y conectados múltiples veces.
